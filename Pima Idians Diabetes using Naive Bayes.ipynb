# reading the dataset

```python
pima_indians_df = pd.read_csv("/kaggle/input/pima-indians-diabetes-database/diabetes.csv")
pima_indians_df.describe().T
```

```python
pima_indians_df.head(10)
```

```python
names = list(pima_indians_df.columns)
names
```

# How many 0s are there in each column ?

```python
def check_zeros(names):
    for i in names:
        if 0 not in pima_indians_df[i]: continue
        try:
            print("{:.<30s}{:d}".format(i, pima_indians_df[i].value_counts()[0]))
        except KeyError:
            print("{:.<30s}{:d}".format(i, 0))
check_zeros(names)
```

# Data cleaning

```python
import random
for i in names:
    if i == 'Pregnancies' or i == 'Age' or i == 'Outcome'or i == 'DiabetesPedigreeFunction': continue
        
    middle_value = int(pima_indians_df.describe()[i].loc['50%'])
    most_value = int(pima_indians_df.describe()[i].loc['75%'])
    try:
        pima_indians_df[i] = pima_indians_df[i].replace(0,random.randrange(middle_value, most_value))
        print(i)
    except ValueError:
        print(middle_value, most_value)
```

```python
pima_indians_df.head(10)
```

```python
check_zeros(names)
```

```python
array = pima_indians_df.values
X = array[:,0:8] 
Y = array[:,8]
test_size = 0.15
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=test_size, random_state=7)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, Y_train)
print(model)
```

```python
expected = Y_test
predicted = model.predict(X_test)

print(metrics.confusion_matrix(expected, predicted))
print(metrics.classification_report(expected, predicted))
```
